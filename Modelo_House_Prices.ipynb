##### Identificação dos Outliers por meio do método IQR (Intervalo Interquartil)

###### O IQR (Interquartile Range), ou intervalo interquartil, é uma medida estatística de dispersão que representa a diferença entre o terceiro quartil (Q3, ou 75º percentil) e o primeiro quartil (Q1, ou 25º percentil) de um conjunto de dados.
###### O multiplicador 1.5 no cálculo do IQR é usado para identificar outliers ao estabelecer limites além dos quais os dados são considerados extremos. Ele define uma faixa que cobre a maioria dos dados legítimos, mesmo em distribuições assimétricas, é uma escolha empírica e equilibrada.

```python
# Selecionar as variáveis numéricas, excluindo 'SalePrice'
numerical_columns = df_train.select_dtypes(include=['float64', 'int64']).columns.tolist()  # Seleciona as colunas numéricas
numerical_columns = [col for col in numerical_columns if col != 'SalePrice']  # Excluir 'SalePrice'

# Função para substituir os outliers pelo limite calculado com IQR
def replace_outliers_with_limits_IQR(df, column):
    # Calcular o primeiro e terceiro quartil
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Calcular o IQR
    IQR = Q3 - Q1
    
    # Definir os limites superior e inferior
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Substituir os outliers pelos limites
    df[column] = np.where(df[column] < lower_limit, lower_limit, df[column])
    df[column] = np.where(df[column] > upper_limit, upper_limit, df[column])
    
    return df

# Aplicar a função para cada coluna numérica, exceto 'SalePrice'
for col in numerical_columns:
    df_train = replace_outliers_with_limits_IQR(df_train, col)

# Verificar as primeiras linhas após o tratamento
df_train.head()
```

###### Caso os valores ultrapassem os limites definidos, a substituição do mesmo pelo limite correspondente é realizada.

##### Aplicação da transformação BoxCox

###### A transformação Box-Cox é uma técnica estatística utilizada para estabilizar a variância e tornar os dados mais próximos de uma distribuição normal. É particularmente útil quando os dados apresentam assimetrias ou heterocedasticidade (variação não constante dos erros).

###### Quando Utilizar a Transformação Box-Cox: Use quando os dados apresentam assimetria ou heterocedasticidade, especialmente em análises que assumem normalidade ou variância constante.
###### Evite seu uso se: Os dados contêm valores negativos ou zero, e a transformação necessária antes de aplicar Box-Cox pode distorcer os resultados.
###### As relações não lineares nos dados são fundamentais para o problema e não devem ser eliminadas.
###### Se Box-Cox não for aplicável devido a restrições nos dados, alternativas como transformação de logaritmo, raiz quadrada ou Yeo-Johnson podem ser consideradas.

```python
# Aplicando Box-Cox (somente para valores positivos)
df_train['SalePrice'], _ = boxcox(df_train['SalePrice'])
```

```python
# Plotar o histograma da variável transformada
sns.histplot(df_train['SalePrice'], kde=True, bins=30)
plt.title('Distribuição de SalePrice (Box-Cox)')
plt.xlabel('SalePrice (Box-Cox)')
plt.ylabel('Frequência')
plt.show()

# Verificar a assimetria
new_skewness = skew(df_train['SalePrice'])
print(f'Assimetria após Box-Cox: {new_skewness:.2f}')
```

##### Aplicar a transformação BoxCox para variáveis com assimetria maior do que 1 e, caso a variável tenha valores negativos, aplicaremos a transformação Yeo-Johnson.

```python
# Calcular a assimetria das variáveis numéricas
skewness = df_train[numerical_columns].apply(skew)

# Filtrar variáveis com alta assimetria (|skew| > 1)
high_skew = skewness[abs(skewness) > 1].index
print(f'Variáveis com alta assimetria:\n{high_skew}')

# Criar um objeto do PowerTransformer para Yeo-Johnson
pt = PowerTransformer(method='yeo-johnson')

# Dicionário para armazenar os lambdas de Box-Cox (caso queira reverter)
boxcox_lambdas = {}

for col in high_skew:
    # Verificar se a variável contém apenas valores positivos
    if (df_train[col] > 0).all():  # Para Box-Cox, todos os valores devem ser positivos
        df_train[col], lambda_ = boxcox(df_train[col])
        boxcox_lambdas[col] = lambda_  # Salvar o lambda de Box-Cox
        print(f'Box-Cox aplicado na variável: {col} com lambda: {lambda_}')
    else:
        # Para variáveis com valores negativos ou zero, aplica-se Yeo-Johnson
        df_train[col] = pt.fit_transform(df_train[[col]])
        print(f'Yeo-Johnson aplicado na variável: {col}')
```

```python
new_skewness = df_train[high_skew].apply(skew)
print(f'Nova assimetria após transformação:\n{new_skewness}')
```

##### 3. Correlação de Pearson

###### A Correlação de Pearson é uma medida estatística que avalia a força e a direção da relação linear entre duas variáveis numéricas. O valor da correlação de Pearson varia entre -1 e 1, na qual os limites indicam correlação forte (positiva ou negativa).

##### 3.1 Teste ANOVA para variáveis categóricas

###### O teste ANOVA (Analysis of Variance) é uma técnica estatística usada para comparar as médias de três ou mais grupos e determinar se há diferenças estatisticamente significativas entre eles. Ele verifica se a variação entre as médias dos grupos é maior do que a variação dentro dos próprios grupos, levando em conta a variância geral.
###### As variáveis independentes precisam ser categóricas e a dependente contínua.

```python
# Selecionar as variáveis categóricas do df_train
categorical_columns = df_train.select_dtypes(include=['object']).columns

# Criar um dicionário para armazenar os resultados
anova_results = {}

# Loop para realizar o teste ANOVA para cada variável categórica
for col in categorical_columns:
    # Agrupar os dados por categoria e obter as distribuições de 'SalePrice'
    grouped = df_train.groupby(col)['SalePrice']
    
    # Aplicar o teste ANOVA
    f_stat, p_value = stats.f_oneway(*[group[1].values for group in grouped])
    
    # Armazenar o p-valor no dicionário de resultados
    anova_results[col] = p_value

# Converter os resultados em DataFrame para visualização
anova_results_df = pd.DataFrame(list(anova_results.items()), columns=['Variable', 'p-value'])

# Exibir as variáveis com p-valor menor que 0.05
significant_vars = anova_results_df[anova_results_df['p-value'] < 0.05]

print("Variáveis com associação significativa com SalePrice:")
print(significant_vars)
```

```python
# Exibir as variáveis com p-valor maior ou igual a 0.05
non_significant_vars = anova_results_df[anova_results_df['p-value'] >= 0.05]

print("Variáveis sem associação significativa com SalePrice:")
print(non_significant_vars)
```

##### 4. Data Splitting: Treino, Validação e Teste

###### Data splitting (divisão de dados) é o processo de dividir o conjunto de dados em duas ou mais partes para permitir a construção e avaliação de modelos preditivos de maneira mais robusta e imparcial. A ideia é usar uma parte dos dados para treinar o modelo e outra parte para testá-lo, a fim de verificar sua capacidade de generalizar para novos dados.

```python
# Definindo o dataframe df_train como exemplo (substitua pelo seu dataset)
# Exemplo de separação entre variáveis dependente e independente
X = df_train.drop('SalePrice', axis=1)  # Selecionando as variáveis independentes
y = df_train['SalePrice']  # A variável dependente

# Divisão de treino e teste (80% treino, 20% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Verificar as dimensões após a divisão
print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
```

##### 4.1 One Hot Encoder


###### O One-Hot Encoder é uma técnica de pré-processamento usada para converter variáveis categóricas em variáveis numéricas binárias. Ele transforma cada valor único de uma variável categórica em uma nova coluna, onde o valor é representado como 1 (verdadeiro) ou 0 (falso), dependendo da presença ou ausência da categoria.

```python
# Selecionar as variáveis categóricas
categorical_columns = X_train.select_dtypes(include=['object']).columns

# Criar um pré-processador para aplicar OneHotEncoding nas variáveis categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), categorical_columns)
    ],
    remainder='passthrough'  # Deixar as variáveis numéricas como estão
)

# Aplicando o OneHotEncoder nos dados de treino e teste
X_train_encoded = preprocessor.fit_transform(X_train)
X_test_encoded = preprocessor.transform(X_test)

# Verificar as dimensões após o OneHotEncoder
print("Dimensões de X_train após OneHotEncoder:", X_train_encoded.shape)
print("Dimensões de X_test após OneHotEncoder:", X_test_encoded.shape)
```

##### 4.2 StandardScaler
###### O StandardScaler é uma técnica de normalização de dados utilizada para transformar variáveis numéricas para que elas tenham média 0 e desvio padrão 1. Esse processo é conhecido como padronização. A ideia principal é garantir que as variáveis estejam em uma escala semelhante, evitando que variáveis com magnitudes maiores dominem o modelo de aprendizado.

##### 5. Aplicação de Modelos de Regressão
###### Um modelo de regressão é uma técnica estatística usada para entender e modelar a relação entre uma variável dependente (ou alvo) e uma ou mais variáveis independentes (ou preditoras). O objetivo é prever o valor da variável dependente com base nos valores das variáveis independentes.

```python
# Definir os modelos
models = {
    'Linear Regression': LinearRegression(),
    'ElasticNet': ElasticNet(),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}
# Armazenar os resultados
results = []

# Testar cada modelo
for name, model in models.items():
    # Treinar o modelo
    model.fit(X_train_encoded, y_train)
    
    # Fazer previsões
    y_pred = model.predict(X_test_encoded)
    
    # Calcular as métricas
    r2 = r2_score(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    
    # Armazenar os resultados
    results.append([name, r2, rmse, mae])

# Criar um DataFrame com os resultados
results_df = pd.DataFrame(results, columns=['Model', 'R²', 'RMSE', 'MAE'])

# Mostrar os resultados
print(results_df)
```

##### 5.1 Método K-Fold
###### O K-Fold Cross Validation é uma técnica de validação de modelos de aprendizado de máquina. Ela divide o dataset em K subconjuntos (folds), e o modelo é treinado e testado K vezes. A cada iteração, um subconjunto é utilizado como conjunto de teste, enquanto os outros K−1 subconjuntos são usados para treino. O processo é repetido para todos os folds, e o desempenho do modelo é avaliado pela média das métricas obtidas em cada iteração.

```python
# Garantir que X_train_encoded e y_train sejam arrays numpy
X_train_encoded = np.array(X_train_encoded)
y_train = np.array(y_train)

# Definir os modelos
models = {
    'Linear Regression': LinearRegression(),
    'ElasticNet': ElasticNet(),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}

# Armazenar os resultados
results = []

# Definir o número de folds
kf = KFold(n_splits=10, shuffle=True, random_state=42)  # 5 folds

# K-Fold Cross Validation
for name, model in models.items():
    fold_results = {'Model': name, 'R²': [], 'RMSE': [], 'MAE': []}
    
    # Iterar sobre os folds
    for train_index, val_index in kf.split(X_train_encoded):
        # Dividir os dados em treino e validação para cada fold
        X_train_fold, X_val_fold = X_train_encoded[train_index], X_train_encoded[val_index]
        y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]
        
        # Treinar o modelo
        model.fit(X_train_fold, y_train_fold)
        
        # Fazer previsões
        y_pred = model.predict(X_val_fold)
        
        # Calcular as métricas
        r2 = r2_score(y_val_fold, y_pred)
        rmse = mean_squared_error(y_val_fold, y_pred, squared=False)
        mae = mean_absolute_error(y_val_fold, y_pred)
        
        # Armazenar as métricas para o fold
        fold_results['R²'].append(r2)
        fold_results['RMSE'].append(rmse)
        fold_results['MAE'].append(mae)
    
    # Calcular as médias das métricas para o modelo
    results.append([
        name,
        np.mean(fold_results['R²']),
        np.mean(fold_results['RMSE']),
        np.mean(fold_results['MAE'])
    ])

# Criar um DataFrame com os resultados
results_df = pd.DataFrame(results, columns=['Model', 'R²', 'RMSE', 'MAE'])

# Mostrar os resultados
print(results_df)
```

##### 5.2 Método LOO
###### O Leave-One-Out Cross Validation (LOO) é uma variação do K-Fold Cross Validation onde o número de folds é igual ao número de observações no dataset. Em cada iteração, o modelo é treinado com todas as observações, exceto uma, que é usada como conjunto de teste. Esse processo é repetido para cada observação no conjunto de dados.

```python
# Definir os modelos
models = {
    'Linear Regression': LinearRegression(),
    'ElasticNet': ElasticNet(),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}

# Armazenar os resultados
results = []

# Criar uma instância do LeaveOneOut
loo = LeaveOneOut()

# Testar cada modelo
for name, model in models.items():
    # Armazenar as métricas para este modelo
    model_results = []
    
    # Realizar a validação Leave-One-Out
    for train_index, val_index in loo.split(X_train_encoded):
        # Dividir os dados em treino e validação
        X_train_fold, X_val_fold = X_train_encoded[train_index], X_train_encoded[val_index]
        y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]
        
        # Treinar o modelo
        model.fit(X_train_fold, y_train_fold)
        
        # Fazer previsões
        y_pred = model.predict(X_val_fold)
        
        # Calcular as métricas
        r2 = r2_score(y_val_fold, y_pred)
        rmse = mean_squared_error(y_val_fold, y_pred, squared=False)
        mae = mean_absolute_error(y_val_fold, y_pred)
        
        # Armazenar as métricas para cada iteração do LOO
        model_results.append([r2, rmse, mae])
    
    # Calcular a média das métricas
    mean_rmse = np.mean([result[1] for result in model_results])
    mean_mae = np.mean([result[2] for result in model_results])
    
    # Armazenar os resultados finais para o modelo
    results.append([name, mean_rmse, mean_mae])

# Criar um DataFrame com os resultados
results_df = pd.DataFrame(results, columns=['Model', 'Mean RMSE', 'Mean MAE'])

# Mostrar os resultados
print(results_df)
```

##### 5.3 Otimização de Hiperparâmetros para a Regressão Ridge (K-Fold)

##### GridSearch

```python
# Garantir que X_train_encoded e y_train sejam arrays numpy
X_train_encoded = np.array(X_train_encoded)
y_train = np.array(y_train)

# Definir o modelo Ridge
ridge_model = Ridge()

# Definir o número de folds
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Definir o espaço de hiperparâmetros para o alpha do Ridge
param_grid = {
    'alpha': np.logspace(-10, 10, 21)  # Explorar uma faixa de valores para o alpha
}

# Configurar a busca em grid
grid_search = GridSearchCV(estimator=ridge_model, param_grid=param_grid, cv=kf, scoring='neg_mean_squared_error', n_jobs=-1)

# Ajustar o modelo usando o GridSearchCV
grid_search.fit(X_train_encoded, y_train)

# Obter os resultados do GridSearch
grid_results = grid_search.cv_results_

# Mostrar os melhores parâmetros e resultados
print(f"Melhor alpha (GridSearch): {grid_search.best_params_['alpha']}")
print(f"Melhor Score (GridSearch): {grid_search.best_score_}")

# Calcular as métricas de desempenho no conjunto de dados
model_grid_ridge_cv = grid_search.best_estimator_

r2_grid = r2_score(y_train, model_grid_ridge_cv.predict(X_train_encoded))
rmse_grid = mean_squared_error(y_train, model_grid_ridge_cv.predict(X_train_encoded), squared=False)
mae_grid = mean_absolute_error(y_train, model_grid_ridge_cv.predict(X_train_encoded))

print(f"GridSearch - R²: {r2_grid}, RMSE: {rmse_grid}, MAE: {mae_grid}")
```

##### Optuna

```python
# Definir a função de objetivo para o Optuna
def objective(trial):
    # Definir o espaço de busca para o alpha usando suggest_loguniform
    alpha = trial.suggest_loguniform('alpha', 1e-10, 1e10)  # Exemplo de distribuição log-uniforme

    # Criar o modelo Ridge com o valor de alpha
    model = Ridge(alpha=alpha)
    
    # Calcular o erro usando K-Fold Cross Validation
    score = cross_val_score(model, X_train_encoded, y_train, cv=kf, scoring='neg_mean_squared_error').mean()
    
    return score  # Retornar o valor negativo da pontuação do erro quadrático médio

# Criar o estudo do Optuna
study = optuna.create_study(direction='maximize')

# Executar a otimização
study.optimize(objective, n_trials=50)

# Exibir os melhores parâmetros encontrados
print(f"Melhor alpha (Optuna): {study.best_params['alpha']}")
print(f"Melhor Score (Optuna): {study.best_value}")

# Ajustar o modelo com o melhor parâmetro encontrado
model_optuna_cv = Ridge(alpha=study.best_params['alpha'])
model_optuna_cv.fit(X_train_encoded, y_train)

# Calcular as métricas de desempenho no conjunto de dados
r2_optuna = r2_score(y_train, model_optuna_cv.predict(X_train_encoded))
rmse_optuna = mean_squared_error(y_train, model_optuna_cv.predict(X_train_encoded), squared=False)
mae_optuna = mean_absolute_error(y_train, model_optuna_cv.predict(X_train_encoded))

print(f"Optuna - R²: {r2_optuna}, RMSE: {rmse_optuna}, MAE: {mae_optuna}")
```
